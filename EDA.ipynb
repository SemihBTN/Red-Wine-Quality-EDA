```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("13-WineQT.csv")
```

```python
korrelasyon = df.corr()
korrelasyon_matrisi = sns.heatmap(korrelasyon , annot=True , cmap="coolwarm" , fmt=".2f" , linewidth=5)
plt.title("Tüm Değişkenlerin Birbirleriyle İlişkileri")
plt.show()
```

```python
alcohol = df[["alcohol","quality"]]
korelasyon_matrisii = alcohol.corr()
sns.heatmap(korelasyon_matrisii , annot=True)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df['fixed acidity'], kde=True, ax=axes[0], color='red')
axes[0].set_title('Fixed Acidity')

sns.histplot(df['volatile acidity'], kde=True, ax=axes[1], color='blue')
axes[1].set_title('Volatile Acidity')

sns.histplot(df['citric acid'], kde=True, ax=axes[2], color='green')
axes[2].set_title('Citric Acid')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.boxplot(df['fixed acidity'], ax=axes[0], color='red')
axes[0].set_title('Fixed Acidity')

sns.boxplot(df['volatile acidity'], ax=axes[1], color='blue')
axes[1].set_title('Volatile Acidity')

sns.boxplot(df['citric acid'], ax=axes[2], color='green')
axes[2].set_title('Citric Acid')

plt.tight_layout()
plt.show()
```
